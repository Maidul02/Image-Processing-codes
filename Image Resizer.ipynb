```python
import os
import PIL
from PIL import Image
```

```python
main_path = "E:/images"

for file in os.listdir(main_path):
    img_path = os.path.join(main_path,file)
    print(img_path)
    print(file)
    img = Image.open(img_path)
    if img.mode != 'RGB':
        img = img.convert('RGB')
    img = img.resize((300,300))
    img.save(str(img_path))
```
